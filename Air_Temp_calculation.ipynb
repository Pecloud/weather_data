```python
import sys
import pandas as pd
```

### read the file and add date

```python
df = pd.read_csv('data/chicago_beach_weather.csv')
```

```python
df['date'] = pd.to_datetime(df['Measurement Timestamp']).dt.strftime('%m/%d/%Y')

df['date']
```

```python
df['Measurement Timestamp'] = pd.to_datetime(df['Measurement Timestamp'])
```

```python
df = df.sort_values(['Station Name', 'Measurement Timestamp'])
```

```python
df.head(50)
```

### caculate air temperature group by name and date

```python
df = df.groupby(['Station Name', 'date']).agg({'Air Temperature': ['min', 'max', 'first', 'last']})
```

```python
for i_row in df.iterrows():
    print(f"{i_row[0][0]},{i_row[0][1]},{i_row[1][('Air Temperature', 'min')]},{i_row[1][('Air Temperature', 'max')]},{i_row[1][('Air Temperature', 'first')]},{i_row[1][('Air Temperature', 'last')]}\n")
```
